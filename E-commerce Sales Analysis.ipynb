```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample e-commerce sales data
data = {
    "Product": ["Shoes", "Shirt", "Shoes", "Watch", "Shirt", "Shoes", "Watch", "Shirt"],
    "Category": ["Fashion", "Fashion", "Fashion", "Accessories", "Fashion", "Fashion", "Accessories", "Fashion"],
    "Price": [2000, 1200, 2000, 5000, 1200, 2000, 5000, 1200],
    "Quantity": [2, 3, 1, 1, 2, 4, 2, 1]
}

df = pd.DataFrame(data)

# Create revenue column
df["Revenue"] = df["Price"] * df["Quantity"]

df
```

```python
revenue_by_product = df.groupby("Product")["Revenue"].sum()
print(revenue_by_product)
```

```python
revenue_by_category = df.groupby("Category")["Revenue"].sum()
print(revenue_by_category)
```

```python
revenue_by_product.plot(kind="bar")
plt.title("Revenue by Product")
plt.ylabel("Revenue")
plt.show()
```

```python
top_product = revenue_by_product.idxmax()
print("Top earning product:", top_product)
```
